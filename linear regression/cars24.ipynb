```python
# make a plot : scatter plot:
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.title('MaxPrice vs Price')
sns.scatterplot(data=data,x='Max Price', y='Selling Price');
```

```python
def estimate_charge(maxprice, w, b):
    return w* maxprice +b
```

```python
# randomaly choose:
w=50
b=100
```

```python
maxprice=data['Max Price']
maxprice
```

```python
estimated_charge= estimate_charge(maxprice,w,b)
```

```python
target=data['Selling Price']	
plt.figure(figsize=(10,6))
plt.plot(maxprice,estimated_charge, 'r' , alpha=0.9);
plt.scatter(maxprice,target,s=8,alpha=0.8);
plt.xlabel('MaxPrice')
plt.ylabel('Price')
plt.legend(['Estimate', 'Actual'])
```

```python
def try_parameters(w,b):
    maxprice=data['Max Price']
    target=data['Selling Price']

    estimated_charges=estimate_charge(maxprice,w,b)
    plt.figure(figsize=(10,6))
    plt.plot(maxprice,estimated_charges,'r',alpha=0.9);
    plt.scatter(maxprice,target,s=8,alpha=0.8);
    plt.xlabel('MaxPrice')
    plt.ylabel('Price')
    plt.legend(['Estimate','Actual']);
```
